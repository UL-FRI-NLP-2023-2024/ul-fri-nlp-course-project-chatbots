```python
from datasets import Dataset
import pandas as pd

boolq_train_path = "nlp/slovene_superglue/SuperGLUE-HumanT/csv/BoolQ/train.csv"
boolq_val_path = "nlp/slovene_superglue/SuperGLUE-HumanT/csv/BoolQ/val.csv"
boolq_test_path = "nlp/slovene_superglue/SuperGLUE-HumanT/csv/BoolQ/test.csv"

data_boolq_train = pd.read_csv(boolq_train_path)
data_boolq_val = pd.read_csv(boolq_val_path)
data_boolq_test = pd.read_csv(boolq_test_path)

train_dataset = Dataset.from_pandas(data_boolq_train)
eval_dataset = Dataset.from_pandas(data_boolq_val)
test_dataset = Dataset.from_pandas(data_boolq_test)

hub_path = "tjasad/Slovene_SuperGLUE_BoolQ"

train_dataset.push_to_hub(hub_path, split="train", private=True)
eval_dataset.push_to_hub(hub_path, split="eval", private=True)
```
