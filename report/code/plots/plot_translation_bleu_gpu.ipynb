```python
import matplotlib.pyplot as plt
import numpy as np


# Data for plotting
batch_sizes = ["FT", "LoRa", "LoHa"]
boolq_1 = [0.405, 0.356, 0.346]
# Bar width
bar_width = 0.6

# Positions of the bars on the x-axis
r1 = np.arange(len(batch_sizes))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Create the plot
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
fig.subplots_adjust(hspace=0.4)

axs[0].grid(axis='y', linestyle='--', linewidth=0.7, zorder=0)
axs[0].bar(r1, boolq_1, color=['#b056d1',  '#f3b864', '#5a7df4'], width=bar_width, label='FT', zorder=3)

for i in range(len(r1)):
    axs[0].text(r1[i], boolq_1[i] + 0.01, str(boolq_1[i]), ha='center', va='bottom', rotation=45, fontsize=6, fontweight='bold')

axs[0].set_xticks([r for r in range(len(batch_sizes))])
axs[0].set_xticklabels(batch_sizes)
axs[0].set_ylabel('Bleu score')
axs[0].set_ylim(0, 0.45)
axs[0].set_xlabel('a', fontsize=14)


cb_1 = [6.4, 8.2, 6.0]


axs[1].grid(axis='y', linestyle='--', linewidth=0.7, zorder=0)
axs[1].bar(r1, cb_1, color=['#b056d1','#f3b864', '#5a7df4'], width=bar_width, label='FT', zorder=3)

for i in range(len(r1)):
    axs[1].text(r1[i], cb_1[i] + 0.01, str(cb_1[i]), ha='center', va='bottom', rotation=45, fontsize=6, fontweight='bold')


axs[1].set_xticks([r for r in range(len(batch_sizes))])
axs[1].set_xticklabels(batch_sizes)
axs[1].set_ylabel('GPU memory (GB)')
axs[1].set_xlabel('b', fontsize=14)


fig.text(0.5, -0.06, 'PEFT', ha='center', va='center', fontsize=14)
fig.text(0.94, 0.5, '.', color="white", ha='center', va='center', fontsize=14)
fig.text(0.07, 0.5, '.', color="white", ha='center', va='center', fontsize=14)
fig.suptitle('Machine translation (Slovenian to English)', fontsize=16)
plt.savefig('translation_metrics.pdf', bbox_inches='tight')
plt.show()
```
